```python
import requests
from bs4 import BeautifulSoup
# import exa_py # make sure exa is installed and configured
import json

# === Step 1: Run exa search ===
def run_exa_search(user_instruction):
    result = exa.search(
        user_instruction,
        type="auto",
        use_autoprompt=True
    )
    return result

# === Step 2: Extract URLs from search results ===
def get_urls_from_results(result):
    urls = []
    for r in result.results:   # exa returns results as list
        if hasattr(r, "url") and r.url:
            urls.append(r.url)
    return urls

# === Step 3: Fetch & clean text ===
def extract_text(url):
    try:
        response = requests.get(url, timeout=10, headers={"User-Agent": "Mozilla/5.0"})
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")

        # remove useless tags
        for tag in soup(["script", "style", "noscript"]):
            tag.decompose()

        text = soup.get_text(separator="\n")
        # clean extra spaces/newlines
        text = "\n".join(line.strip() for line in text.splitlines() if line.strip())
        return text
    except Exception as e:
        return f"âŒ Error fetching {url}: {e}"

# === Step 4: Combine everything ===
def get_combined_content(user_instruction):
    search_results = run_exa_search(user_instruction)
    urls = get_urls_from_results(search_results)

    combined = {}
    for url in urls:
        combined[url] = extract_text(url)

    return combined

# === Usage ===
user_instruction = "Errana Telugu poet biography"   # example
final_response = get_combined_content(user_instruction)

# # Save as JSON
# with open("exa_combined.json", "w", encoding="utf-8") as f:
#     json.dump(final_response, f, ensure_ascii=False, indent=2)

# Preview
for url, text in final_response.items():
    print("="*80)
    print("URL:", url)
    print("="*80)
    print(text[:1000], "...\n")
```
